```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0.001,1,50,endpoint=False)
```

```python
cov = np.zeros((X.shape[0],X.shape[0]))

for i,x in enumerate(X):
    for j,y in enumerate(X):
        cov[i,j] = min(x,y) -x*y  
```

```python
L = np.linalg.cholesky(cov)
n = np.size(X)
m = 10

u = np.random.normal(size=(n, m))
brownian_b_sample = np.dot(L,u)
```

```python
plt.plot(X, brownian_b_sample)

plt.show()
```
